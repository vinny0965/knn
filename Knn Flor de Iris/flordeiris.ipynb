### Selecionando as "features" para classificaÃ§Ã£o em um array NUMPY

### Selecionando as classes para uma classificaÃ§Ã£o em um array NUMPY

# Importando o KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

# Criando um Classificador

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

# Treinando o classificador

```python
knn.fit(x,y)
```

# Predizendo um tipo de flor iris

## Passar o vetor com as caracteristicas da flor

```python
knn.predict([[3.0,3.1,3.0,2.0]])
```

# Acuracia

```python
print(knn.score(x,y))
```
